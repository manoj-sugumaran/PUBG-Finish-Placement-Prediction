### Based on the information in the variable descriptions as well as the correlation of maxPlace to winPlacePerc, I think filling in the missing values for winPlacePerc with 0 seems appropriate.


```python
train_ori['winPlacePerc'] = train_ori['winPlacePerc'].fillna(0)
```

```python
match_type = train_ori.pop('matchType')
Id = train_ori.pop('Id')
groupId = train_ori.pop('groupId')
matchId = train_ori.pop('matchId')
```

```python
X, y = train_ori.drop('winPlacePerc', axis= 1), train_ori['winPlacePerc']
X_train, X_holdout, y_train, y_holdout = train_test_split(X,y, random_state= 17, test_size = 0.3)
```

```python
tree = DecisionTreeRegressor(random_state = 17)
tree.fit(X_train, y_train)
```

```python
pred_holdout = tree.predict(X_holdout)
mean_absolute_error(y_holdout, pred_holdout)
# mean_absolute_error = 0.0823
# Goal is to achieve 0.01
```
